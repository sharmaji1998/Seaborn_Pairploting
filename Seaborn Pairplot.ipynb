## Seaborn Pairplot

```python
import matplotlib.pyplot as plt
import seaborn as sb
import pandas as pd
```

```python
df=sb.load_dataset('tips')
```

```python
df.head()
```

```python
sb.pairplot(df)
plt.show()
```

```python
sb.pairplot(df,hue='sex')
plt.show()
```

```python
sb.pairplot(df,hue='day')
plt.show()
```

```python
sb.pairplot(df,hue='time')
plt.show()
```

```python
sb.pairplot(df,hue='smoker')
plt.show()
```

```python
df=sb.load_dataset('iris')
df.head()
```

```python
sb.pairplot(df)
plt.show()
```

```python
# Set corner=True to plot only the lower triangle
```

```python
sb.pairplot(df,corner=True)
plt.show()
```

```python
sb.pairplot(df,hue='species')
plt.show()
```

```python
# If apne according names tag chaiye toh use markers
```

```python
sb.pairplot(df,hue='species',markers=['*','o','s'])
plt.show()
```

```python
# If only choice k colums ka banana hoto use vars
```

```python
sb.pairplot(df,hue='species',vars=['sepal_length','sepal_width'])
plt.show()
```

```python
sb.pairplot(df,hue='species',x_vars=['sepal_length','sepal_width'],y_vars=['petal_length','petal_width'])
plt.show()
```
